# Pretty 3d plot

```r
x=seq(-pi,pi, length=100)
y=x
f = function(x,y) cos(y)/(1+x^2)
matrix = outer(x,y,f)
matrix
```

```r
contour(x,y,matrix)
```

```r
image(x,y,matrix)
```

```r
persp(x,y,matrix, theta=30)
```
