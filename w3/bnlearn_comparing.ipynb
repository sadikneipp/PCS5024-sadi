## Loading

```r
df = read.table('../datasets/adult/train.csv', header = TRUE, sep = ',')
df
```

## Making columns discrete

```r
df['education_bin'] = factor(df['education_num'] > 9) # if greater than college, TRUE
df['education_bin']
```

```r
hist(as.numeric(df[, 'hr_per_week']), breaks = 100)
#most people answer 40h
```

```r
df['hr_per_week_cat'] = factor(cut(as.numeric(df$hr_per_week),
                               breaks = c(-Inf, 39, 41, Inf), 
                               labels = c("lower", "medium", "upper")))
df['hr_per_week_cat']
```

```r
unique(df['marital_status'])
```

```r
single_idx = df['marital_status'] == 'never-married' |
            df['marital_status'] == 'Divorced'     |
            df['marital_status'] == 'Widowed'     |
            df['marital_status'] == 'Separated'    
df[single_idx, 'marital_status_bin'] = FALSE
df[!single_idx, 'marital_status_bin'] = TRUE
df$marital_status_bin = factor(df$marital_status_bin)
```

## Train-test split of 0.7

```r
chosen_cols = c('income', 'sex', 'marital_status_bin', 'education_bin', 'hr_per_week_cat')

smp_size = floor(0.7 * nrow(df))
## set the seed to make your partition reproductible
set.seed(123)
train_ind = sample(seq_len(nrow(df)), size = smp_size)
full_set = df[ , chosen_cols]
train_set = df[train_ind, chosen_cols]
test_set = df[-train_ind, chosen_cols]
```

```r
test_score = function(fit, set){
    predicted = predict(fit, node = "income", data = set)
    ts_test = ts(predicted)
    truth = ts(set[, "income"])
    return (accuracy(f=ts_test, x=truth))
    }
```

```r
cfit = custom.fit(net, dist = list(sex = cpt_sex, income = cpt_income, education_bin = cpt_edu,
                                   marital_status_bin = cpt_ms, hr_per_week_cat = cpt_hrweek))
cfit
```

```r
test_score(cfit, test_set)
```

61% Error rate
