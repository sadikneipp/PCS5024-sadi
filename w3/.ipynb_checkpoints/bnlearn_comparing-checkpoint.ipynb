## Loading

```r
df = read.table('../datasets/adult/train.csv', header = TRUE, sep = ',')
df
```

## Making columns discrete

```r
df['education_bin'] = factor(df['education_num'] > 9) # if greater than college, TRUE
df['education_bin']
```

```r
hist(as.numeric(df[, 'hr_per_week']), breaks = 100)
#most people answer 40h
```

```r
df['hr_per_week_cat'] = factor(cut(as.numeric(df$hr_per_week),
                               breaks = c(-Inf, 39, 41, Inf), 
                               labels = c("lower", "medium", "upper")))
df['hr_per_week_cat']
```

```r
unique(df['marital_status'])
```

```r
single_idx = df['marital_status'] == 'never-married' |
            df['marital_status'] == 'Divorced'     |
            df['marital_status'] == 'Widowed'     |
            df['marital_status'] == 'Separated'    
df[single_idx, 'marital_status_bin'] = FALSE
df[!single_idx, 'marital_status_bin'] = TRUE
df$marital_status_bin = factor(df$marital_status_bin)
```

```r
male_idx = df['sex'] == 'Male'
df[male_idx, 'sex_bin'] = FALSE
df[!male_idx, 'sex_bin'] = TRUE
df$sex_bin = factor(df$sex_bin)
```

```r
df['sex_bin']
```

## Train-test split of 0.7

```r
chosen_cols = c('income', 'sex', 'marital_status_bin', 'education_bin', 'hr_per_week_cat')

smp_size = floor(0.7 * nrow(df))
## set the seed to make your partition reproductible
set.seed(123)
train_ind = sample(seq_len(nrow(df)), size = smp_size)

train_set = df[train_ind, chosen_cols]
test_set = df[-train_ind, chosen_cols]
```

## bnlearn hand-crafted net

```r
library(bnlearn)
library(forecast)
net = empty.graph(chosen_cols)
```

```r
arc.set = matrix(c("education_bin", "income", "marital_status_bin", "income", "education_bin", "marital_status_bin",
                  "education_bin", "hr_per_week_cat", "hr_per_week_cat", "income", "sex", "income"),
          ncol = 2, byrow = TRUE,
          dimnames = list(NULL, c("from", "to")))
arc.set
```

```r
arcs(net) = arc.set
plot(net)
```

```r
net
```

```r
cpt_edu = matrix(c(0.25, 0.75),ncol =2, dimnames = list(NULL, "education_bin" = c("TRUE", "FALSE")))
cpt_sex = matrix(c(0.5, 0.5), ncol = 2, dimnames = list(NULL, "sex" =  c("Female", "Male")))

cpt_ms = matrix(c(0.4, 0.6, 0.7, 0.3))
dim(cpt_ms) = c(2, 2)
dimnames(cpt_ms) = list("marital_status_bin" = c("TRUE", "FALSE"), "education_bin" = c("TRUE", "FALSE"))

cpt_hrweek = matrix(c(0.4, 0.3, 0.3, 0.3, 0.2, 0.5))
dim(cpt_hrweek) = c(3, 2)
dimnames(cpt_hrweek) = list("hr_per_week_cat" = c("lower", "medium", "upper"), "education_bin" =  c("TRUE", "FALSE"))

cpt_income = matrix(c(0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8, 0.4, 0.6, 0.7, 0.3, 0.2, 0.8))
dim(cpt_income) = c(2, 2, 2, 2, 3)
dimnames(cpt_income) = list("income" = c("<=50K", ">50K"), "sex" = c("Female", "Male"),
                        "marital_status_bin" = c("TRUE", "FALSE"), 
                        "education_bin" =  c("TRUE", "FALSE"),
                        "hr_per_week_cat" = c("lower", "medium", "upper"))
```

```r
dfit = custom.fit(net, dist = list(income = cpt_income, sex_bin = cpt_sex, education_bin = cpt_edu, marital_status_bin  = cpt_ms,
                                   hr_per_week_cat = cpt_hrweek))
dfit
```

```r
predict(dfit, node= "income", data= test_set[1, ])
```

```r
test_score(dfit)
```

```r
fit = bn.fit(net, train_set , debug = TRUE)
fit
```

```r
test_score = function(fit){
    predicted = predict(fit, node = "income", data = test_set)
    ts_test = ts(predicted)
    truth = ts(test_set[, "income"])
    return (accuracy(f=ts_test, x=truth))
    }
```
